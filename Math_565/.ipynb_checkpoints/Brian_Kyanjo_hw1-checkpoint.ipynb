# N0.1

```python
# Starting Condition: 
kmax = 20
a=5

x = zeros((kmax,1))
x[0] = a  # intial guess

tol = 1e-12 #tolerance

for k in range(0,kmax-1):
    x[k+1] = 0.5*(x[k]  + a/x[k])
    if abs(x[k+1]-x[k]) < tol:
        max_k = k+1
        break        
    
set_printoptions(precision=16)
print(x[:max_k])
```

```python
#function f(x)
def f(x):
    return (1/(1+x**2))


def Trapeziodal(f,a,b,n):
    h=(b-a)/n #step size
    y=np.zeros(n+1)
    
    for i in range(1,n):
        xi=a+i*h
        y[i]=f(xi)
    
    Tn=0.5*h*(f(a)+2*(sum(y)) +f(b)) #Trapeziodal rule
    return Tn
```

```python
Trapeziodal(f,a=0,b=1,n=10) # for n=10
```

# (c)

```python
def polynomial(a, xo): # a is an array of coefficients of the polynomial
    power=xo
    p=a[0]+ a[1]*xo
    for i in range(2,len(a)):
        power = power*xo
        p=p+a[i]*power
    return p
    
    
```

# Test

```python
a=[4,4.7,0,-1,pi,3.1]
xo=1
polynomial(a,xo)
```

# Plot 

```python
xo=linspace(-2,2)
Poly=polynomial(a,xo)
plot(xo,Poly)
ylabel('p(x)')
xlabel('x')
title('A graph of p(x) against x')
show()
```
