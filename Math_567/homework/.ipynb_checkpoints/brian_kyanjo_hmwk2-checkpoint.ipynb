# 1.a)

```python
#function u(t)
def u(t):
        return (exp(cos(5*(t-0.1))))
    
    
#computing samples of uj
def uj(N):
    t=[(2*pi*j)/N for j in range(N)] ; tj=array(t)
    uj=u(tj) #uj samples
    return uj

def uprime(uj,N):
    '''
    Arguments
    ---------
    N: Total number of iterations
    uj: Samples of u(t) at j=0,1,...,N
    
    Return
    ------
    uprim: real part of the inverse fast fourier tansform of dk
    '''
    #apply fft
    ucap=fft(uj)
    n=int((N-1)/2)
    #computing dk, j is complex
    dk1=[]
    dk2=[]
    for k in range(0,n+1):
        dk_1=j*k*ucap[k]
        dk1.append(dk_1)
    for k in range(-n,0):
        dk_2=j*k*ucap[k]
        dk2.append(dk_2)
        
    dk=dk1 + dk2
    uprim=real(ifft(dk))
    return uprim
```

# Using N=63

```python
#N=63
N=63
tj6=zeros(N)
for j in range(N): #for N odd
    tj6[j]=(2*pi*j)/N
    
uj6=u(tj6)
up63=uprime(uj6,N)
print('up63=',up63)
```

```python
N=127
tj1=zeros(N)
for j in range(N): #for N odd
    tj1[j]=(2*pi*j)/N
    
uj1=u(tj1)
#N=127
up127=uprime(uj1,N)
print('up127=',up127)
```

```python
def up(t):
    return (-5*sin(5*(t-0.1))*u(t))

#for N=63
uexact=up(tj6)
Error63=uexact-up63
#for N=127
uexact=up(tj1)
Error127=uexact-up127
```

```python
figure(1)
plot(tj6,Error63)
title('for n=63')
ylabel('Error')
xlabel('tj')
show()
```

```python
figure(2)
plot(tj1,Error127)
title('for n=127')
ylabel('Error')
xlabel('tj')
show()
```

# c

```python
def uapp(uj,N):
    h=(2*pi)/N
    DN=zeros((N,N))
    DN[N-1,N-2]=-1
    DN[N-1,0]=1
    for j in range(0,N-1):
        DN[j,j]=0
        DN[j,j-1]=-1
        DN[j,j+1]=1
    DN1=(1/(2*h))*DN
    up=matmul(DN1,transpose(uj))
    return(up)
```

```python
#for N=63
N=63
uapp6=uapp(uj6,N=63)
#for N=127
N=127
uapp1=uapp(uj1,N=127)

#error
#for N=63
uexact=up(tj6)
Error6=uexact-uapp6
#for N=127
uexact=up(tj1)
Error1=uexact-uapp1
```

```python
figure(3)
plot(tj6,Error6)
title('for n=63')
ylabel('Error')
xlabel('tj')
show()
```

```python
figure(4)
plot(tj1,Error1)
title('for n=127')
ylabel('Error')
xlabel('tj')
show()
```

# using fft

```python
N=64
sig=0.1
def a(t):
    return((1/(sig*sqrt(2*pi)))*(exp((-0.5*(t**2))/sig**2)))
def x(t):
    return(1+0.05*cos(16*t))*cos(2*t)
tl=zeros(N)
for l in range(N):
    tl[l]=-pi + ((2*pi*l)/N)

#compute the DFTs of a and x
a=fft(a(tl))
x=fft(x(tl))

#Obtain b 
b=[]
for m in range(N):
    b.append(a[m]*x[m])
#Compute the inverse
bcap=real(ifft(b))
print('b=',bcap)
```
