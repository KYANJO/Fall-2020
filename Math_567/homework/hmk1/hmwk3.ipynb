```python
def fd2tpbvp(p,q,r,alpha,beta,m):
    a=0;b=1
    h=(b-a)/(m+1)
    x=zeros(m)
    x[m-1]=1
    A=zeros((m,m))
    A[m-1,m-1]=(-2/(h**2))-q(b)
    F=zeros(m)
    rh=zeros(m)
    rh[m-1]=r(b)
    e1=zeros(m);em=zeros(m)
    e1[0]=1;em[m-1]=1
    for j in range(m-1):
        x[j]=a+j*h
        A[j,j]=(-2/(h**2))-q(x[j])
        A[j,j+1]=(1/(h**2))-((p(x[j]))/(2*h))
        A[j+1,j]=(1/(h**2))+p(x[j])
        rh[j]=r(x[j])
        f1=(1/(h**2))+(p(x[j]))/(2*h)
        f2=(1/(h**2))-(p(x[j]))/(2*h)
    
    F=rh-(f1*alpha*e1)-(f2*beta*em)
    U_h=linalg.solve(A,F) #inverse(A)*F
    return U_h   
```

```python
#numerically
p=lambda x: 2*(tan(x))
q=lambda x: 0
r=lambda x: 2
alpha=0
beta=0

uapprox=[]
m=[10,20,40,80,160]
for j in m:
    uapp=fd2tpbvp(p,q,r,alpha,beta,j)
    uapprox.append(uapp)
print('Numerical approximation of TPBVP for m=10 is: ',uapprox[0])
print('\nNumerical approximation of TPBVP for m=20 is: ',uapprox[1])
print('\nNumerical approximation of TPBVP for m=40 is: ',uapprox[2])
print('\nNumerical approximation of TPBVP for m=80 is: ',uapprox[3])
print('\nNumerical approximation of TPBVP for m=160 is: ',uapprox[4])
```

```python
#exact solution
uexact=lambda x: (x-1)*(tan(x))
Uexact=[]
h=[]
m=[10,20,40,80,160]
for j in m:
    xj=zeros(j)
    for i in range(1,j):
        hi=1/(j+1)
        xj[i]=i*hi
    h.append(hi)
    u_exact=uexact(xj)
    Uexact.append(u_exact)
    print('\nExact solution for m=',j,'is', u_exact)
```
